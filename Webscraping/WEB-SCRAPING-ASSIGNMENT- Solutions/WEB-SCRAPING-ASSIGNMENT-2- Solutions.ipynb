## Q6: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field.
You have to scrape 4 attributes of each sneaker:
1. Brand
2. Product Description
3. Price
4. Discount

```python
driver = webdriver.Chrome(r'C:\Users\bhanu\chromedriver.exe')
driver.maximize_window()

url = 'https://www.flipkart.com/'
driver.get(url)

search_job = driver.find_element(By.CLASS_NAME,"_3704LK")
search_job.send_keys('sneakers')

search_btn = driver.find_element(By.CLASS_NAME,"_34RNph")
search_btn.click()

brand = []
summary = []
price= []
discount=[]

for page in range(1,4):
    url = 'https://www.flipkart.com/search?q=sneakers&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(page)
    driver.get(url)
    for i in driver.find_elements(By.CLASS_NAME, "_2WkVRV"):
        brand.append(i.text)
    for i in driver.find_elements(By.CLASS_NAME,'IRpwTa'):
        summary.append(i.text)
    for i in driver.find_elements(By.CLASS_NAME, "_30jeq3"):
        price.append(i.text)
    price = price[:len(price)-5]
    
    box=[]
    tot = driver.find_elements(By.CLASS_NAME, '_2B099V')
    for i in tot:
        box.append(i.text)
    for i in range(0,40):
        if('%' in box[i]):
            discount.append(box[i].split('%')[0][-2:])
        else:
            discount.append('null')
```

```python
SNK = pd.DataFrame({'Brand':brand, 'Summary':summary,'Price':price,'Discount':discount})
SNK
```

## Q8: Go to webpage https://www.amazon.in/
### Enter “Laptop” in the search field and then click the search icon.
### Then set CPU Type filter to “Intel Core i7”

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time

driver = webdriver.Chrome(r'C:\Users\bhanu\chromedriver.exe')
driver.maximize_window()

url = 'https://www.amazon.in/'
driver.get(url)

search_product = driver.find_element(By.XPATH, "//div[@class='nav-search-field ']//input")
search_product.send_keys('Laptop')


search_btn = driver.find_element(By.CLASS_NAME,"nav-right")
search_btn.click()

i7 = driver.find_element(By.XPATH,"//li[@aria-label='Intel Core i7']//div[@class='a-checkbox a-checkbox-fancy s-navigation-checkbox aok-float-left']")

i7.click()

title =[]
ratings = []
price=[]


for i in driver.find_elements(By.XPATH, '//span[@class="a-size-medium a-color-base a-text-normal"]'):
    title.append(i.text)

title = title[:10]

for i in driver.find_elements(By.CLASS_NAME, "a-price-whole"):
    price.append(i.text)
price=price[:10]

for i in driver.find_elements(By.XPATH, '//div[contains(@class, "s-result-item s-asin")]'):
    ratings_box = i.find_elements(By.XPATH, './/div[@class="a-row a-size-small"]/span')

    if ratings_box != []:
        rating = ratings_box[0].get_attribute('aria-label')
    else:
        rating = 'null'
    
    ratings.append(rating)

ratings = ratings[:10]

LAP = pd.DataFrame({'Title':title, 'Price':price, 'Ratings':ratings}) 
LAP
```

## Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpage https://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quote

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time

driver = webdriver.Chrome(r'C:\Users\bhanu\chromedriver.exe')
driver.maximize_window()

url = 'https://www.azquotes.com/'
driver.get(url)



menu = driver.find_elements(By.XPATH,'//div[@class="mainmenu"]/ul/li')
topqt_btn = menu[4]
topqt_btn.click()

title=[]

author =[]

tags= []

for i in driver.find_elements(By.XPATH, "//ul[@class='list-quotes']/li/div/p/a[@class='title']"):
    title.append(i.text)

for i in driver.find_elements(By.XPATH, "//ul[@class='list-quotes']/li/div/div/a"):
    author.append(i.text)

for i in driver.find_elements(By.CLASS_NAME, 'tags'):
    tags.append(i.text)

TQ = pd.DataFrame({"Title":title, 'Author':author, 'Type_of_Quote':tags})
TQ
```

## Q10: Write s python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpage https://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make the DataFrame

```python
driver = webdriver.Chrome(r'C:\Users\bhanu\chromedriver.exe')
driver.maximize_window()

url = 'https://www.jagranjosh.com/'
driver.get(url)

menu = driver.find_elements(By.XPATH,'//div[@class="hashtag"]//ul/li')
gk_btn = menu[8]
gk_btn.click()

menu = driver.find_elements(By.XPATH,'//div[@id="popluarGK"]/ul/li')
pm_btn = menu[1]
pm_btn.click()

box = driver.find_elements(By.XPATH, '//div[@class="table-box"]/table/tbody/tr/td')


box = box[:90]

pm =[]
born_death = []
terms=[]
dur=[]
remarks=[]

import numpy as np

for i in np.arange(1,90,5):
    pm.append(box[i].text)    

for i in np.arange(2,90,5):
    born_death.append(box[i].text)

for i in np.arange(3,90,5):
    terms.append(box[i].text.split('\n')[0])

for i in np.arange(3,85,5):
    dur.append(box[i].text.split('\n')[1])
dur.append('present')

for i in np.arange(4,90,5):
    remarks.append(box[i].text)

PML = pd.DataFrame({"Name":pm, 'Born_Dead':born_death,'Terms':terms,'Duration':dur, 'Remarks':remarks})
PML
```
