```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
driver = webdriver.Chrome(r'C:\Users\bhanu\chromedriver.exe')
driver.maximize_window()

url = 'https://www.youtube.com/watch?v=kX3nB4PpJko'
driver.get(url)

time.sleep(5)
```

```python
comments = []
```

```python
for i in driver.find_elements(By.XPATH, '//div[@class="style-scope ytd-comment-renderer"]/div[@class="style-scope ytd-comment-renderer"]/ytd-expander/div/yt-formatted-string'):
    comments.append(i.text)
```

```python
times = []
```

```python
for i in driver.find_elements(By.XPATH, '//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]/a'):
    times.append(i.text)
```

```python
df = pd.DataFrame({'Comments':comments, 'Time Commented':times})
df
```
